## 4 Modelo Linear usando o Scikit Learn

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X, y)
```

## 5 Coeficientes

```python
#Escrever a fórmula apenas
```

```python
model.intercept_
```

```python
model.coef_
```

## 6 Descobrindo y_pred

```python
y_pred = model.predict(X)
y_pred
```

## 7 Cálculo da métrica MSE

### Usando Scikit Learn

```python
mse = metrics.mean_squared_error(y, y_pred)
mse
```

### Usando somente Python

```python
mse_python = np.sum((y - y_pred) ** 2) / len(y)
mse_python
```

## 8 Modelo de Regressão Linear para a feature que melhor se ajusta a função linear

```python
model_melhor = linear_model.LinearRegression()
```

```python
X2_ = X2.reshape((442,1))
```

```python
model_melhor.fit(X2_,y)
```

- Cálculo dos coeficientes de forma manual

```python
y_ = [[i] for i in y]
```

```python
# B1
b1 = np.sum((X2_ - np.mean(X2_)) * (y_ - np.mean(y_))) / np.sum((X2_ - np.mean(X2_)) ** 2)
b1
```

```python
# B0
b0 = np.mean(y_) - b1 * np.mean(X2_)
b0
```

- Cálculo dos coeficientes de forma automatizada

```python
# B0
model_melhor.intercept_
```

```python
# B1
model_melhor.coef_
```

- Gráfico de dispersão 

```python
y_pred_melhor = model_melhor.predict(X2_)
```

```python
plt.grid()
plt.scatter(X2_, y, color='red')
plt.plot(X2_, y_pred_melhor)
```

- Cálculo do MSE

```python
mse_melhor = metrics.mean_squared_error(y, y_pred_melhor)
mse_melhor
```
