```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
x = np.random.rand(20)
x = (x*100).round(2)
x = np.resize(x, (20,1))
```

```python
x
```

```python
def normalize(X):
    X_norm = np.copy(X)
    n_cols = X.shape[1]
    for i in range(n_cols):
        X_norm[:, i] = (X[:, i] - np.min(X[:, i])) / (np.max(X[:, i]) - np.min(X[:, i]))
    return X_norm
```

```python
b = normalize(x)
b
```

```python
plt.hist(b, bins=20);
```

```python
sns.distplot(b, bins=20, rug=True);
```

```python
def standardize(X):
    X_std = np.copy(X)
    n_cols = X.shape[1]
    for i in range(n_cols):
        X_std[:, i] = (X[:, i] - np.mean(X[:, i])) / np.std(X[:, i])
    return X_std
```

```python
c = standardize(x)
c
```

```python
plt.hist(c, bins=20);
```

```python
sns.distplot(c, bins=20, rug=True);
```
